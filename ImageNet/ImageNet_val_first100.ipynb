```python
import os
import shutil
from tqdm import tqdm

src = "/home/ecbm4040/WAWJ-convmixer/pytorch-image-models/ImageNet1k-val"  

dst = "/home/ecbm4040/WAWJ-convmixer/pytorch-image-models/ImageNet_val_first100"
os.makedirs(dst, exist_ok=True)

classes = sorted(os.listdir(src))[:100]

print(f"Total classes selected: {len(classes)}")

for cls in tqdm(classes, desc="Copying classes"):
    src_cls = os.path.join(src, cls)
    dst_cls = os.path.join(dst, cls)

    if not os.path.exists(dst_cls):
        shutil.copytree(src_cls, dst_cls)
    else:
        for fname in os.listdir(src_cls):
            src_file = os.path.join(src_cls, fname)
            dst_file = os.path.join(dst_cls, fname)
            if not os.path.exists(dst_file):
                shutil.copy2(src_file, dst_file)

print("Done! ")
print(f"Saved to: {dst}")
```
